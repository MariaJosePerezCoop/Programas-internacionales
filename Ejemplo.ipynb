```python
#procesar y hacer un layout (espacio en pag web)
app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Styling dataset by using html components', style = {'textAlign':'center',\
                                            'marginTop':20,'marginBottom':20}),

        dcc.Dropdown( id = 'dropdown',
        options = [
            {'label':'Promedio', 'value':'Promedio'},
            {'label': 'Puntaje asignado', 'value':'PuntajeAsignación'},
            {'label': 'Materias totales', 'value':'Núm. MatTotales'}
            ],
        value = 'Promedio'),
        dcc.Graph(id = 'bar_plot')
    ])
```

```python
#crear funcion para establecer entradas y salidas de datos (función para crear y actualizar el gráfico)
#control y gráfico asociado al control
@app.callback(Output(component_id='bar_plot', component_property= 'figure'),
              [Input(component_id='dropdown', component_property= 'value')])

def graph_update(dropdown_value):
    print(dropdown_value)
    fig = go.Bar(x = df['Nivel'], y = df['{}'.format(dropdown_value)],\
                     line = dict(color = 'firebrick', width = 4))
                     
    
    fig.update_layout(title = 'Puntuación por nivel',
                      xaxis_title = 'Nivel',
                      yaxis_title = 'Calificación'
                      )
    return fig


```

```python
#ejecutar aplicación
app.run_server(mode='inline')
```
